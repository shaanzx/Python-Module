```python
import requests

url = "https://api.github.com/events"

response = requests.get(url)

if response.status_code == 200:
    print("Success")
    if 'application/json' in response.headers.get('Content-Type', ''):
        data = response.json()
        print(data)
    
    print(response.text)
else:
    print(f"Error while fetching {response.status_code}")
```

```python
import requests

url = 'https://httpbin.org/post'
data = {'username' : 'user', 'password' : 'pass'}

response = requests.post(url, data=data)

if response.status_code == 200:
    print('Data retrive successfully done!')
    print(response.text)
    print(response.url)
else:
    print(f'Faild to login {response.status_code}')
```

```python
# URL Parameters

import requests

url = 'https://jsonplaceholder.typicode.com/posts'
parms = {'userId' : 1, 'id' : 2}

response = requests.get(url, params=parms)
if response.status_code == 200:
    print('Successfully Done!')
    print(response.text)
    print(response.url)
else: 
    print('Data feching error!')
```

```python
import requests

url = 'https://httpbin.org/get'
username = 'user'
password = 'pass'

response = requests.get(url, auth=(username, password))
if response.status_code == 200:
    print('Authenticate success!')
    print(response.text)
else:
    print(f'Data feching error! {response.status_code}')
```
